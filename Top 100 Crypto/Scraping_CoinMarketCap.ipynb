```python
import requests
import pandas as pd


api_key = 'API'
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'

# Menentukan parameter yang ingin digunakan
parameters = {
    'start': '1',
    'limit': '100',
    'convert': 'USD'
}

# Menentukan header dengan API key
headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': api_key,
}

# Mengambil data dari API
response = requests.get(url, headers=headers, params=parameters)
data = response.json()

# Membuat DataFrame dari data yang diambil
df = pd.DataFrame(data['data'], columns=['name', 'quote', 'cmc_rank', 'last_updated'])
df['current_price'] = df['quote'].apply(lambda x: x['USD']['price'])
df['price_change_percentage_24h'] = df['quote'].apply(lambda x: x['USD']['percent_change_24h'])
df['market_cap'] = df['quote'].apply(lambda x: x['USD']['market_cap'])
df['total_volume'] = df['quote'].apply(lambda x: x['USD']['volume_24h'])

# Memilih kolom yang diinginkan
df = df[['name', 'current_price', 'price_change_percentage_24h', 'market_cap', 'total_volume']]

# Menyimpan DataFrame ke file CSV
df.to_csv('crypto_data_coinmarketcap.csv', index=False)

print("Data telah disimpan ke 'crypto_data_coinmarketcap.csv'")
```
